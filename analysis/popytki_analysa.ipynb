```python
# for ABC analysis

for df in dfs:
    y = int('20' + df['Sale Date'].iloc[0].split('/')[-1])  # строка с годом таблицы
    df['y'] = [y]*len(df)
```

```python
def change(x):
    if str(x[1]) in x[0]:
        x[0] = x[0].replace(str(int(x[1])), 'this year')
    if str(x[1] + 1) in x[0]:
        x[0] = x[0].replace(str(int(x[1] + 1)), 'next year')
    if str(x[1] - 1) in x[0]:
        x[0] = x[0].replace(str(int(x[1] - 1)), 'previous year')
    return x[0]

for df in dfs:
    df['Item Name'] = df[['Item Name', 'y']].apply(change, axis=1)
```
